```python

url = 'https://www.spe.org/datasets/dataset_1/csv_files/dataset_1_all_wells/production_data'
tempfile = os.path.join(os.getcwd(), 'temp.csv')
r = requests.get(url)
print(r.content)
with open(tempfile, 'wb') as file:
    file.write(r.content)
df = pd.read_csv(tempfile)
```
